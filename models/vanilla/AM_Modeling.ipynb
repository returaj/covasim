```python
data_path = '/mnt/d/books/iitm/agentBased/data/tn/DDW-3300C-13.xls'
```

```python
data = pd.read_excel(data_path)
```

```python
header, clean_data = data[:6], data[6:]
```

```python
print(clean_data.columns)
clean_data.head()
```

```python
clean_data = clean_data.rename(columns={'Unnamed: 0':'TableName', 'Unnamed: 1':'StateCode', 'Unnamed: 2': 'DistCode', 'Unnamed: 3':'AreaName', 'C-13 SINGLE YEAR AGE RETURNS BY RESIDENCE AND SEX ':'Age', 'Unnamed: 5':'TotalPersons', 'Unnamed: 6': 'TotalMales', 'Unnamed: 7': 'TotalFemales', 'Unnamed: 8': 'RuralPersons', 'Unnamed: 9': 'RuralMales', 'Unnamed: 10': 'RuralFemales', 'Unnamed: 11':'UrbanPersons', 'Unnamed: 12': 'UrbanMales', 'Unnamed: 13': 'UrbanFemales'})
```

```python
clean_data = clean_data.reset_index(drop=True)
clean_data.head()
```

```python
chennai_data = clean_data[clean_data['AreaName'] == 'District - Chennai (02)'].reset_index(drop=True)
```

```python
 chennai_data.head()
```

```python
def get_age_group(data, grp_size=5):
    age_grp = {};
    start_age = 0; cnt = 0
    for end_age in range(100):
        tp = int(data[data['Age'] == end_age]['TotalPersons'])
        cnt += tp
        if start_age + grp_size -1 == end_age:
            age_grp[f'{start_age}-{end_age}'] = cnt
            start_age = end_age + 1; cnt = 0
    age_grp['100+'] = int(data[data['Age'] == '100+']['TotalPersons'])
    return age_grp
```

```python
chennai_age_grp = get_age_group(chennai_data)
x = chennai_age_grp.keys()
y = [chennai_age_grp[k] for k in x]

plt.rcParams["figure.figsize"] = [15.50, 5.50]
plt.rcParams["figure.autolayout"] = True
plt.bar(x, y)
plt.xlabel('chennai age distribution')
plt.ylabel('age count')
```
